```python
def lowercase_strip_reed_techno(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(' ','').split(',')[-1]
        return sentence
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''
    
def lowercase_strip_indeed(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(' ','').split('-')[-1]
        return sentence
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''

def lowercase_strip_24(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(' ','').split(',')[0]
        return sentence
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''
    
def lowercase_strip_today1(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('Posted', '').replace('Salary', '').replace('\r', '').replace('\t', '').replace('\n', '').replace('Recruiter','').strip().split(',')
        return sentence[0]
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''
    
def lowercase_strip_today2(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
#         sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(' ','').split(',')[0]
        return sentence[4]
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''
    
def lowercase_strip_indeed2(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        wanted_list = ['day', 'month', 'year','months', 'days']
        sentence = [word for word in sentence if 'day' in word or 'month' in word ]
        if len(sentence) > 0:
            result = sentence[0].replace('-','').strip()
        else:
            result = 'Not available'
        return result
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''
    
def lowercase_strip_technojob(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\t', '').replace('\n', '').replace('Recruiter','').strip().split(',')
        return sentence[0]
    elif isinstance(sentence, (str,)):
        result = sentence.replace('\n','').strip().lower()
        return result
    else:
        return ''
    
    
def lowercase_strip_reed(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(',','.').replace('..','.').strip()
        return sentence
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''    
    
  
    
def lowercase_strip_reed2(sentence):
    '''
    strip all the non-alphabetical characters
    '''
    if isinstance(sentence, (list,)):
        sentence = ' '.join(sentence).replace('\r', '').replace('\n', '').replace(',','.').replace('..','.').strip().split(' ')[1]
        return sentence
    elif isinstance(sentence, (str,)):
        clean_text = sentence.strip().lower()
        result = re.sub(r"[^a-zA-Z]+", ' ', clean_text)
        return result
    else:
        return ''    
        
    
    
```

```python
STOPWORD = stopwords.words('english') + list(string.punctuation)
```

```python
def clean_discription(descrip_list):
    if isinstance(descrip_list, (list,)):
        descrip_list = ' '.join(descrip_list).replace('\r', '').replace('\n', '').strip()
        clean_descrip =' '.join([i for i in word_tokenize(descrip_list.lower()) if i not in STOPWORD])
    return clean_descrip
```

```python
for i in data_today:
    location_clean = lowercase_strip_today2(i['location'])
    d1 = {'location': location_clean}
    company_clean = lowercase_strip_today1(i['company'])
    d2 = {'company': company_clean}
    title_clean = lowercase_strip_today2(i['title'])
    d3 = {'title': title_clean}
    descrip_clean = clean_discription(i['description'])
    d4 = {'description': descrip_clean}
    salary_clean = lowercase_strip_today1(i['salary'])
    d5 = {'salary': salary_clean}
    posted_clean = lowercase_strip_today1(i['date_listed'])
    d6 = {'date_listed': posted_clean}

    
    i.update(d1)
    i.update(d2)
    i.update(d3)
    i.update(d4)
    i.update(d5)
    i.update(d6)
```
